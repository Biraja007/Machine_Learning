Lets have a visualisation of our dataset:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Lets read our data:

```python
iris = pd.read_csv(r'D:\Data-Sets\Iris.csv')
```

Lets read the top elements in our dataset:

```python
iris.head()
```

Lets find out the number of species:

```python
iris['Species'].value_counts()
```

Lets make a scatterplot of the Iris features:

```python
iris.plot(kind='scatter',x='SepalLengthCm', y='SepalWidthCm')
```

A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure:

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm', data=iris, height=5)
```

Lets see the datatypes present:

```python
iris.info()
```

Now lets use seaborn FacetGrid to color the scatterplot by species:

```python
sns.FacetGrid(iris, hue = 'Species', height=5) \
    .map(plt.scatter, 'SepalLengthCm','SepalWidthCm') \
    .add_legend()
```

Lets look at an individual feature in Seaborn through a boxplot:

```python
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
```

One way we can extend this plot is adding a layer of individual points on top through Seaborn's striplot
 
use jitter=True so that all the points don't fall in single vertical lines

Saving the resulting axes as ax each time causes the resulting plot to be shown on top of the previous axes

```python
ax = sns.boxplot(data=iris, x = 'Species',y = 'PetalLengthCm')
ax = sns.stripplot(data=iris, x='Species', y='PetalLengthCm', jitter=True, edgecolor='green')
```

Another useful seaborn plot is the pairplot, which shows the bivariate relation between each pair of features:

```python
sns.pairplot(iris.drop('Id', axis=1), hue='Species', height=3)
```

Lets make a boxplot with Pandas on each feature split out by species:

```python
iris.drop('Id', axis=1).boxplot(by='Species', figsize=(12,6))
```

```python
sns.factorplot('SepalLengthCm', data=iris, hue='Species', kind='count' )
```
